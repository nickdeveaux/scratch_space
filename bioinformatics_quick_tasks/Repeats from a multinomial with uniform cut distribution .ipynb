A thought experiment from a conversation with Mariano Gabitto

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
n = 40000
length = 40000
pvals = [1 / float(length) for i in range(length)]
```

```python
draws= np.random.multinomial(n, pvals, size=None)
```

```python
draws[draws > 1]
```

```python
np.sum(draws[draws > 1])
```

```python
np.random.multinomial(20, [1/6.]*6, size=1)
```

```python
y = []
for i in range(n):
    draws= np.random.multinomial(i, pvals, size=None)
    y.append(np.sum(draws[draws > 1]))
```

```python
plt.figure(figsize=(15,10))
plt.plot((y))
plt.title('Number of duplicates as a function of cells (assuming an 100% efficiency and 1/40000 probability)')
plt.xlabel('Number of cells')
```
