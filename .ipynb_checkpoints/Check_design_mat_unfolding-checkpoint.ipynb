```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
import seaborn as sns
import pandas as pd
```

```python
num = 1000
```

```python
g1 = np.random.normal(size=num)
```

```python
g2 = np.random.normal(loc = 0.5, size=num)
```

```python
data = np.concatenate((g1, g2))
```

```python
plt.scatter(data, np.random.normal(loc = 0.0, size=num*2))
```

```python
condition = [ 0 if i < num else 1 for i in range(num * 2)]
```

```python
batch = np.random.randint(2, size = 2*num)
```

```python
train_labels
```

```python

# Create linear regression object
regr = linear_model.LinearRegression()

train_data = np.array(data).reshape(-1, 1)
train_labels = np.stack([condition, batch]).transpose()
# train_labels = np.stack([df['nb_and_ny'], df['b_and_y'], df['b_and_ny'], df['nb_and_y']]).transpose()
# Train the model using the training sets
regr.fit(train_data, train_labels)

# Make predictions using the testing set
diabetes_y_pred = regr.predict(train_data)
```

```python
diabetes_y_pred
```

```python

# The coefficients
print('Coefficients: \n', regr.coef_)
# The mean squared error
print("Mean squared error: %.2f"
      % mean_squared_error(train_labels, diabetes_y_pred))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % r2_score(train_labels, diabetes_y_pred))

# Plot outputs
plt.scatter(train_data, condition,  color='black')

plt.plot(train_data, diabetes_y_pred, color='blue', linewidth=3)
```

```python
df = pd.DataFrame.from_dict({'x':data, 'y':condition, 'b':batch})
```
