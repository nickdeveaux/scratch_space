```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set(color_codes=True)
from scipy.stats import bernoulli
```

## Initial Conditions

We start by generating x and y data from a joint gaussian distribution with a covariance of 0.6. The data matrix is clipped so that all entries are greater than zero.

```python
# Generate initial covariance matrix
mean = 3
size = 10000
covariance = [[1.0, 0.6], [0.6, 1.0]]
data = np.clip(np.random.multivariate_normal([mean, mean], covariance, size), 0, 10)
```

```python
x = [dat[0] for dat in data]
y = [dat[1] for dat in data]
df = pd.DataFrame({'x':x, 'y':y})
```

```python
g = sns.jointplot(x="x", y="y", data=df, kind="kde", color="m")
g.plot_joint(plt.scatter, c="w", s=2, linewidth=1, marker="+")
g.ax_joint.collections[0].set_alpha(0)
g.set_axis_labels("$X$", "$Y$");
```

```python
p = 0.95
# Generate Bernoulli random variates
r_x = bernoulli.rvs(p, size=size)
r_y = bernoulli.rvs(p, size=size)
```

```python
# If the Bernoilli trial was a success, set the original value to 0
df['x_dropped'] = df['x']
df['x_dropped'][r_x > 0] = 0
df['y_dropped'] = df['y']
df['y_dropped'][r_y > 0] = 0
```

```python
g = sns.jointplot(x="x", y="y_dropped", data=df, kind="kde", color="m")
g.plot_joint(plt.scatter, c="w", s=2, linewidth=1, marker="+")
g.ax_joint.collections[0].set_alpha(0)
g.set_axis_labels("$X$", "$Y$");
```

### Plot Non-zero entries 

Although the majority of the data is now at (0,0), we can find the non-zero entries and recover the original relationship between x and y

```python
filtered_df = df[(df['x'] > 0) & (df['y_dropped'] > 0)]
```

```python
g = sns.jointplot(x="x", y="y_dropped", data=filtered_df, kind="kde", color="m")
g.plot_joint(plt.scatter, c="w", s=2, linewidth=1, marker="+")
g.ax_joint.collections[0].set_alpha(0)
g.set_axis_labels("$X$", "$Y$");
```

#### Conclusion
By independently setting 80% of the entries in x and y to zero, the pearson R has remained high, but the power has decreased due to the loss of >95% of the data points. The p-value for this correlation finding is only ~ 10^-6. 

```python
# Generate Bernoulli random variates
r_x = bernoulli.rvs(p, size=size)
r_y = bernoulli.rvs(p, size=size)
```

```python
# If the Bernoilli trial was a success, multiply the original value by a uniform random variable in [0, 1] 
df['x_lowered'] = df['x']
df['x_lowered'][r_x > 0] = df['x'][r_x > 0] * np.random.rand(len(df['x'][r_x > 0]))
df['y_lowered'] = df['y']
df['y_lowered'][r_x > 0] = df['y'][r_x > 0] * np.random.rand(len(df['y'][r_x > 0]))
```

```python
g = sns.jointplot(x="x_lowered", y="y_lowered", data=df, kind="kde", color="m")
g.plot_joint(plt.scatter, c="w", s=2, linewidth=1, marker="+")
g.ax_joint.collections[0].set_alpha(0)
g.set_axis_labels("$X$", "$Y$");
```

#### Conclusion
By independently setting 80% of the entries in x and y to a fraction of their original value (according to a uniform distribution from the original data value to 0), the pearson R has dropped to 0.42, but the power has not decreased as much: the p-value for this correlation is still very low, on the order of 10^-42

## Dropout via exponential function

The paper I am trying to replicate states that their simulated dropout was "induced by subsampling
from an exponential distribution such that 80% of values are set to 0."

```python
arb_scale = .05
exp_x = np.random.exponential(size=size, scale=arb_scale)
exp_y = np.random.exponential(size=size, scale=arb_scale)
```

```python
len(exp_x[exp_x < 0.08])
```

```python
exp_x[exp_x < 0.08] = 0
exp_y[exp_y < 0.08] = 0
```

```python
df['exp_x'] = df['x'] * exp_x
df['exp_y'] = df['y'] * exp_y
```

```python
plt.hist(exp_y, bins=50, )
```

```python
g = sns.jointplot(x="exp_x", y="exp_y", data=df, kind="kde", color="m")
g.plot_joint(plt.scatter, c="b", s=2, linewidth=1, marker="+")
g.ax_joint.collections[0].set_alpha(0)
g.set_axis_labels("$X$", "$Y$");
```

#### Conclusion
By independently setting 80% of the entries in x and y to 0, and the rest to a fraction (as drawn from a clipped exponential distribution) of their original value, the correlation is lost

## Dropout via exponential function: Take 2

This time, set the dropout to 0 via an independent Bernouilli trial

```python
arb_scale = .2
size = 10000
exp_x = np.random.exponential(size=size, scale=arb_scale)
exp_y = np.random.exponential(size=size, scale=arb_scale)
```

```python
p = 0.8
# Generate Bernoulli random variates
r_x = bernoulli.rvs(p, size=size)
r_y = bernoulli.rvs(p, size=size)
```

```python
# Modify all values by the exponential distribution random variable
# If the Bernoilli trial was a success, set the original value to zero
df['exp_x_and_bernouilli'] = df['x'] * exp_x
df['exp_x_and_bernouilli'][r_x > 0] = 0 
df['exp_y_and_bernouilli'] = df['y'] * exp_y
df['exp_y_and_bernouilli'][r_y > 0] = 0 
```

```python
g = sns.jointplot(x="exp_x_and_bernouilli", y="exp_y_and_bernouilli", data=df, kind="kde", color="m")
g.plot_joint(plt.scatter, c="b", s=2, linewidth=1, marker="+")
g.ax_joint.collections[0].set_alpha(0)
g.set_axis_labels("$X$", "$Y$");
```

#### Conclusion
By independently setting 80% of the entries in x and y to 0, and the rest to a fraction (as drawn from an exponential distribution, unclipped) of their original value, the correlation is further lost in this example

## Coin flip Check

```python
flips = 1000000
p = 0.5
other_p = 0.2
results = []
for i in range(flips):
    tmp_p = p
    if np.random.rand() > p:
        tmp_p = other_p
    if np.random.rand() > tmp_p:
        results.append(0)
    else:
        results.append(1)
```

```python
heads = [i for i in results if  i == 1]
```

```python
len(heads)
```

```python
# expected value of heads:
expected_heads = 0.5 * 0.5 + 0.5 * other_p
```

```python
expected_heads
```

```python
analytic_variance = expected_heads * (1 - expected_heads)
```

```python
analytic_variance
```
